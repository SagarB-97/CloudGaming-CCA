```python
import plotly.graph_objects as go
import pandas as pd
```

```python
flow_grind_df = pd.read_csv('fg.csv')
df_0 = flow_grind_df[flow_grind_df['flow'] == 0]
df_1 = flow_grind_df[flow_grind_df['flow'] == 1]

fig = go.Figure()
fig.add_trace(go.Scatter(x = df_0['time'], 
                         y = df_0['thru(Mbps)'].ewm(span = 50).mean(), 
                         name = 'TCP-LP'))
fig.add_trace(go.Scatter(x = df_1['time'], 
                         y = df_1['thru(Mbps)'].ewm(span = 50).mean(), 
                         name = 'Reverse traffic (TCP-LP)'))
fig.update_xaxes(title = 'Time (s)')
fig.update_yaxes(title = 'Throughput (Mbps)')
fig.show()
```

```python
downq_df_lp = pd.read_csv('saved_data/downq.csv')
downq_df_cubic = pd.read_csv('downq.csv')
fig = go.Figure()
fig.add_trace(go.Scatter(x = downq_df_lp['time(sec)'], 
                         y = downq_df_lp['qlen'].rolling(10).mean(), 
                         name = 'LP'))
fig.add_trace(go.Scatter(x = downq_df_cubic['time(sec)'], 
                         y = downq_df_cubic['qlen'].rolling(10).mean(), 
                         name = 'cubic'))
fig.update_xaxes(title = 'Time')
fig.update_yaxes(title = 'Queue length #packets')
```

```python
upq_df_lp = pd.read_csv('saved_data/upq.csv')
upq_df_cubic = pd.read_csv('upq.csv')
fig = go.Figure()
fig.add_trace(go.Scatter(x = upq_df_lp['time(sec)'], 
                         y = upq_df_lp['qlen'].rolling(10).mean(), 
                         name = 'LP'))
fig.add_trace(go.Scatter(x = upq_df_cubic['time(sec)'], 
                         y = upq_df_cubic['qlen'].rolling(10).mean(), 
                         name = 'cubic'))
fig.update_xaxes(title = 'Time')
fig.update_yaxes(title = 'Queue length #packets')
```

```python
flow_grind_df = pd.read_csv('fg.csv')
df_0 = flow_grind_df[flow_grind_df['flow'] == 0]
df_1 = flow_grind_df[flow_grind_df['flow'] == 1]

fig = go.Figure()
fig.add_trace(go.Scatter(x = df_0['time'], 
                         y = df_0['cwnd'].rolling(1).mean(), 
                         name = 'TCP-LP'))
fig.add_trace(go.Scatter(x = df_1['time'], 
                         y = df_1['cwnd'].rolling(10).mean(), 
                         name = 'Reverse traffic (cubic)'))
fig.update_xaxes(title = 'Time (s)')
fig.update_yaxes(title = 'cwnd')
fig.show()
```
